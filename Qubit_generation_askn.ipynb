```python
from qiskit import QuantumCircuit

# Ask user for qubit count
n = int(input("Enter number of qubits to generate: "))

# Create a quantum circuit
qc = QuantumCircuit(n)

# Add operations (optional)
qc.h(0)          # Put qubit 0 in superposition
qc.cx(0, 1)      # Entangle qubit 0 and 1

# Display the circuit
print(qc.draw())
print(f"\nGenerated {n} qubits!")
```
