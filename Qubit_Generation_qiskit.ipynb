```python
from qiskit import QuantumCircuit

# Create a quantum circuit with 2 qubits
qc = QuantumCircuit(2)

# Initialize qubits (default |0вҹ©)
qc.h(0)  # Apply Hadamard gate to create superposition (|0вҹ© вҶ’ (|0вҹ© + |1вҹ©)/вҲҡ2)
qc.cx(0, 1)  # Entangle qubits (CNOT gate)

# Visualize the circuit
print(qc.draw())
```
